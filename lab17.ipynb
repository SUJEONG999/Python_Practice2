```python
from wordcloud import WordCloud            ## 워드 클라우드 모듈을 사용한다 
import matplotlib.pyplot as plt 
```

```python
myfontpath = "data/THEdog.ttf" 
```

```python
import nltk   
filename = 'naverhotel'
inputFileName = 'data/'+filename
data = open(inputFileName+'.txt', 'r', encoding='utf-8').read()
data #출력하여 내용 확인
```

```python
from konlpy.tag import Okt      
t = Okt()
tokens_ko = t.nouns(data)                ## 텍스트에서 명사를 추출한다. 
print(tokens_ko[:10])
```

```python
nouns_text = nltk.Text(tokens_ko)
```

```python
data = nouns_text.vocab().most_common(150)   
```

```python
print(data)
```

```python
wc = WordCloud(font_path=myfontpath,                                    ## .한글에 대한 위치를 표시한다. 
                      relative_scaling = 0.3,
                       width = 700,
                        height = 500,
                      background_color='white',
                      ).generate_from_frequencies(dict(data))             ## .단어별 빈도수를 딕셔너리로 변환해서 전달한다 
plt.figure(figsize=(15,10))
plt.imshow(wc)                                                      ## .이미지를 출력한다 
plt.axis("off")                                                            ## 그래프에 대한 축을 표시하지 않는다 
plt.show()
```
